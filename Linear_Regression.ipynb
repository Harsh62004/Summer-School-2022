```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (13, 10)
import numpy as np
import pandas as pd
```

```python
size = 100

x = np.random.random(size=size) * 100
w_0 = 10
w_1 = 5
y = w_0 + w_1 * x + np.random.random(size=size) * 100
```

```python
plt.scatter(x, y)
plt.ylabel('V (m/s)')
plt.xlabel('T (s)')
```

### Pseudo Inverse

```python
# Prepare x and y
X_nm = np.ones((100, 2))
X_nm[:, 1] = x
X_nm.shape, X_nm
```

```python
# Calculate pseudoinverse
x_1 = np.matmul(X_nm.T, X_nm)
x_2 = np.linalg.inv(x_1)
pseudo_inv = np.matmul(x_2, X_nm.T)

pseudo_inv.shape
```

```python
np.linalg.inv
```

```python
# Calculate w
w = np.matmul(pseudo_inv, y)
```
